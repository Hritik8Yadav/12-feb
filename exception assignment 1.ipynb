```python
# Ans 1:
# Exception are the unwanted event occurs during the normal execution of the programs.
# Exception occurs during the runtime of the program while errors occurs due to the syntax errors.
```

```python
# Ans 2:
# If the exception is not handled then it will disrupt the whole program execution as the python is interpreted.
# so due to this the program will not execute itself if any error is raised. To avoid this we use exception handling.
# for example:
a=1
b="one"
c=a+b
print("This line will not execute due to no exception handling")
# this will raise error due to this the print statement will not be executed
```

```python
# We can use exception handling to avoid this as:
try:
    a=1
    b="one"
    c=a+b
except Exception as e:
    print(e)
finally:
    print("This line will now print due to exception handling")
```

```python
# Ans 3:
# 'try' and except statements are used to catch and handle exceptions in python.
# for example:
try:
    a=1/0
except Exception as e:
    print(e)
```

```python
# Ans 4:
# (a). try:- try is used to make the code try to run if there is no error given by the code.
# (b). else:- else is the block which will only execute only if there is no error given by the code.
# (c). finally:- finally is the block of code which will run independently of error given or not.
# (d). raise:- raise is the keyword which is used to raise the error by defining.

# for example:
def error_raise(time_in_hours):
    if (time_in_hours>24):
        raise Exception("Entered time is more then 24 hours")
    
try:
    time_in_hours=int(input("Enter the time: "))
    error_raise(time_in_hours)
except Exception as e:
    print(e)
else:
    print("the time is : ",time_in_hours)
finally:
    print("program ends here")
```
